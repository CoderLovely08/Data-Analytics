```python
import pandas as pd
import numpy as np
```

```python
sale2015 = pd.read_csv('AdventureWorks_Sales_2015.csv')
sale2016 = pd.read_csv('AdventureWorks_Sales_2016.csv')
sale2017 = pd.read_csv('AdventureWorks_Sales_2017.csv')
salesData = pd.concat([sale2015,sale2016,sale2017])
```

```python
territory = pd.read_csv('AdventureWorks_Territories.csv')
salesTerritory = pd.merge(salesData,territory, left_on = 'TerritoryKey',right_on='SalesTerritoryKey')
```

```python
prod = pd.read_csv('AdventureWorks_Products.csv')
pcs = pd.read_csv('AdventureWorks_Product_Subcategories.csv')
pc = pd.read_csv('AdventureWorks_Product_Categories.csv')
products = pd.merge(pd.merge(prod,pcs),pc)
```

```python
cust = pd.read_csv('AdventureWorks_Customers.csv',encoding='latin1')
customerSale = pd.merge(salesTerritory,cust)
```

```python
output = pd.merge(products,customerSale)
output['TotalSale']=output.ProductPrice*output.OrderQuantity
output['TotalCost']=output.ProductCost*output.OrderQuantity
output['TotalProfit']=output.TotalSale-output.TotalCost
if np.where(output.OrderLineItem>output.OrderQuantity):
    output['EstimatedSale']= (output.OrderLineItem-output.OrderQuantity)*output.ProductPrice
else: output['EstimatedSale'] =0
# output.to_csv('FinalOutput2.csv')
output
```
