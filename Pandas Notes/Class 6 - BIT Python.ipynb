```python
mergedSalesOutput = pd.concat([table1,table2,table3])
mergedSalesOutput.to_csv('FinalSalesData.csv')
import os
print(os.getcwd())
```

```python
productCat = pd.read_csv('AdventureWorks_Product_Categories.csv')
productSubCat = pd.read_csv('AdventureWorks_Product_Subcategories.csv')
```

```python
productCat
productCatWithSub = pd.merge(productCat,productSubCat)
```

```python
territoryDF = pd.read_csv('AdventureWorks_Territories.csv')
territoryDF
```

```python
salesWithTerritory = pd.merge(mergedSalesOutput,territoryDF,left_on='TerritoryKey',right_on='SalesTerritoryKey')
```

```python
salesWithTerritory
```

```python
allProducts = pd.merge(productCatWithSub,pd.read_csv('AdventureWorks_Products.csv'))
allProducts.to_csv('AllProducts.csv')
```

```python
productSalesTerrirtory = pd.merge(allProducts,salesWithTerritory)
```
